```python
# Splitting text into sentences

import pandas as pd

# Read the Excel file
file_path = '/home/pratimathapa/code/PratimaThapa/DisLex_Project/xlsx_files/English/Indonesia.xlsx'
df = pd.read_excel(file_path)

# Create an empty list to store rows
rows_to_add = []

# Flag to check if the previous row was an integer
prev_row_integer = False

# Iterate through each row in the DataFrame
for index, row in df.iterrows():
    text = str(row['Sentence'])  # Convert the value to a string

    # Check if the text is only an integer
    if text.isdigit():
        # Set the flag to indicate the previous row was an integer
        prev_row_integer = True

        # Store the integer in the row for later concatenation
        integer_row = text
    else:
        # If the previous row was an integer, concatenate its text with the current row
        if prev_row_integer:
            text = f"{integer_row} {text}"
            prev_row_integer = False  # Reset the flag

        if '(' in text:
            sections = text.split('(')
            main_text = sections[0].strip()
            subsections = [f'({section.strip()}' for section in sections[1:]]

            # Append the main row with original text before segmentation
            row_copy = row.copy()
            row_copy['Sentence'] = main_text
            rows_to_add.append(row_copy)

            # Append the subsections directly after the main row
            for subsec in subsections:
                new_row = row.copy()
                new_row['Sentence'] = subsec
                rows_to_add.append(new_row)
        else:
            rows_to_add.append(row)

# Create a new DataFrame from the list of rows
new_df = pd.DataFrame(rows_to_add)

# Filter out rows where 'Sentence' column is empty or not a string
new_df = new_df[new_df['Sentence'].astype(str).str.strip() != '']

# Resetting the index
new_df.reset_index(drop=True, inplace=True)

# Display the first few rows of the updated DataFrame
new_df.head(50)
#new_df.tail(50)

# Save the updated dataframe to a new Excel file
#output_file_path = '/home/pratimathapa/code/PratimaThapa/DisLex_Project/xlsx_files/English_edited/Indonesia.xlsx'
#new_df.to_excel(output_file_path, index=False)
```

# Correcting pdf files with spaces between words

```python
import pandas as pd
import wordninja

# Read the CSV file into a pandas DataFrame
data = pd.read_csv('/home/pratimathapa/code/PratimaThapa/DisLex_Project/csv_files/Correction_needed/Republic of Korea.csv')

# Retrieve text from the 'Sentence' column (assuming 'Sentence' is the column name)
text_column = 'Sentence'

# Function to remove spaces and split text into words using Wordninja
def remove_spaces_and_split(text):
    text_without_spaces = ''.join(text.split())
    words_with_spaces = ' '.join(wordninja.split(text_without_spaces))
    return words_with_spaces

# Apply the function to the 'Sentence' column
data[text_column] = data[text_column].apply(remove_spaces_and_split)

# Display a sample of the modified DataFrame to verify
print(data.sample(10))  # Display a sample of 10 rows to check the 'Sentence' column

# Save the updated dataframe to a new Excel file
#output_file_path = '/home/pratimathapa/code/PratimaThapa/DisLex_Project/xlsx_files/English_edited/Republic of Korea.xlsx'
#data.to_excel(output_file_path, index=False)
```

# Extracting text from Sierra Leone pdf link

```python
import pandas as pd
import requests
from io import BytesIO
import PyPDF2
import nltk

# Download NLTK sentence tokenizer if not already downloaded
nltk.download('punkt')

# Fetch the PDF content using requests or any other method
response = requests.get('https://sierralii.gov.sl/akn/sl/act/2020/3/eng@2020-06-11/source.pdf')

# Assuming response.content contains the PDF content
pdf_content = response.content

# Create a BytesIO object and pass the content
pdf_file = PyPDF2.PdfReader(BytesIO(pdf_content))

# Initialize empty list to store sentences
sentences = []

# Extract sentences from each page of the PDF
for page_num in range(len(pdf_file.pages)):
    page = pdf_file.pages[page_num]
    extracted_text = page.extract_text()

    # Remove '\n' and '\t' from the extracted text
    cleaned_text = extracted_text.replace('\n', '').replace('\t', '')

    # Tokenize text into sentences using NLTK
    extracted_sentences = nltk.sent_tokenize(cleaned_text)
    sentences.extend(extracted_sentences)

# Create DataFrame
data = {
    'Sentence': sentences,
    'Language': ['English'] * len(sentences),
    'Country': ['Sierra Leone'] * len(sentences),
}

df = pd.DataFrame(data)

# Displaying the DataFrame
print(df.head())

# Save the updated dataframe to a new Excel file
#output_file_path = '/home/pratimathapa/code/PratimaThapa/DisLex_Project/xlsx_files/English_edited/Sierra Leone.xlsx'
#df.to_excel(output_file_path, index=False)
```

# Extracting Thai text from Disaster Law of Thailand

```python
import pandas as pd
import requests
from io import BytesIO
import PyPDF2
import nltk
import unicodedata

# Download NLTK sentence tokenizer if not already downloaded
nltk.download('punkt')

# Fetch the PDF content using requests or any other method
response = requests.get('https://disasterlaw.ifrc.org/sites/default/files/media/disaster_law/2020-08/Thailand_DMAct_2007.pdf')

# Assuming response.content contains the PDF content
pdf_content = response.content

# Create a BytesIO object and pass the content
pdf_file = PyPDF2.PdfReader(BytesIO(pdf_content))

# Initialize empty list to store sentences
sentences = []

# Extract sentences from each page of the PDF
for page_num in range(len(pdf_file.pages)):
    page = pdf_file.pages[page_num]
    extracted_text = page.extract_text()

    # Remove '\n' and '\t' from the extracted text
    cleaned_text = extracted_text.replace('\n', '').replace('\t', '')

    # Tokenize text into sentences using NLTK
    extracted_sentences = nltk.sent_tokenize(cleaned_text)

    # Check each sentence for Thai characters and only add those to the list
    for sentence in extracted_sentences:
        # Check if the sentence contains Thai characters
        has_thai_chars = any(unicodedata.name(char).startswith('THAI') for char in sentence if char.isprintable())
        if has_thai_chars:
            sentences.append(sentence)

# Create DataFrame
data = {
    'Sentence': sentences,
    'Language': ['Thai'] * len(sentences),
    'Country': ['Thailand'] * len(sentences),
}

df = pd.DataFrame(data)

# Displaying the DataFrame
print(df.tail())

# Save the updated dataframe to a new Excel file
#output_file_path = '/home/pratimathapa/code/PratimaThapa/DisLex_Project/xlsx_files/Thailand_Thai.xlsx'
#df.to_excel(output_file_path, index=False)
```
